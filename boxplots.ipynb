```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/final_analysis_df_boxplot.csv')
df['ROI'] = df['ROI'].str.upper()
# df.replace({'GLAND': ''}, regex=True, inplace=True)
df
```

```python
import matplotlib.ticker as ticker
plot = sns.boxplot(x='ROI', y='DSC', data=df)
plot.autoscale(enable=True, axis='x', tight=True)
sns.set(rc = {'figure.figsize':(35,8)})
plot.set_xlabel("Regions of Interest", fontsize = 20)
plot.set_ylabel("Dice Similarity Coefficient", fontsize = 20)
plot.tick_params(labelsize=15)
plot.set_title("Dice Similarity Coefficient by Organ", fontsize = 30)
plot.figure.savefig('data/final_analysis_boxplot.png')
```
