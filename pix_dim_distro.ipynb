```python
import matplotlib
import seaborn as sns

matplotlib.use('TKAgg')
from matplotlib import pylab as plt
import nibabel as nib
from nibabel import nifti1
from nibabel.viewers import OrthoSlicer3D
import os
print(os.getcwd())
from tqdm import tqdm
```

```python
inputdir ='data/AMOS22/imagesTr/'
x_pix_list = []
y_pix_list = []
z_pix_list = []

for file in tqdm(os.listdir(inputdir)):
    file_path = os.path.join(inputdir, file)
    img = nib.load(file_path)
    x_pix_list.append(img.header['pixdim'][1])
    y_pix_list.append(img.header['pixdim'][2])
    z_pix_list.append(img.header['pixdim'][3])
```

```python
# Plot
kwargs = dict(hist_kws={'alpha':.6}, kde_kws={'linewidth':2})

sns.distplot(x_pix_list, color="dodgerblue", label="x distribution", **kwargs)
sns.distplot(y_pix_list, color="orange", label="y distribution", **kwargs)
sns.distplot(z_pix_list, color="deeppink", label="z distribution", **kwargs)
plt.legend()
plt.show()
```

```python
pair_pix_list = []
for file in tqdm(os.listdir(inputdir)):
    file_path = os.path.join(inputdir, file)
    img = nib.load(file_path)
    x = img.header['pixdim'][1]
    y = img.header['pixdim'][2]
    pair = str(x) + 'x' + str(y)
    pair_pix_list.append(pair)
    # pair_list.append(x + 'x' + y)
pair_pix_list
```
