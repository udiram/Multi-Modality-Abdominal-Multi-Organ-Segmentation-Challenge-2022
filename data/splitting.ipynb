```python
import json
import os
print(os.getcwd())
```

```python
key = json.load(open('data/AMOS22/task1_dataset.json'))
```

```python
image_label_pairs = key['training']
# print(image_label_pair)
dict = image_label_pairs[0]
print(dict)
```

```python
imagename= dict['image'].split('/')[-1]
# firstpartofimagename = imagename.split('.')[1]
# print(firstpartofimagename)
```

```python
import shutil
CT_files = []
os.makedirs('data/CT_files/imagesTr')
for i, file in enumerate(os.listdir('data/AMOS22/imagesTr/')):
    image_label_pairs = key['training']
    inner_dictionary = image_label_pairs[i]
    imagename = inner_dictionary['image'].split('/')[-1]
    new_image_name = 'img' + str(i) + '.nii.gz'
    current_path = 'data/AMOS22/imagesTr/' + file
    new_path = 'data/CT_files/imagesTr/' + new_image_name
    print(imagename)
    print(file)
    if imagename == file:
        shutil.copy(current_path, new_path)
        CT_files.append(new_image_name)
    else:
        print('not matching')
```
